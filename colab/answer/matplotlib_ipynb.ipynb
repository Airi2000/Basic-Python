## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
data1=pd.read_csv("https://raw.githubusercontent.com/Airi2000/Basic-Python/master/colab/quizzes/matplotlib_quiz/hist_data.csv")
```

```python
plt.hist(data1["aデータ"],bins=30,alpha=0.3,label="aデータ")
plt.hist(data1["bデータ"],bins=30,alpha=0.3,color="r",label="bデータ")
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
data2=pd.read_csv("https://raw.githubusercontent.com/Airi2000/Basic-Python/master/colab/quizzes/matplotlib_quiz/scatter_data.csv")
```

```python
d2r=data2.loc[data2["class"]==0]
d2b=data2.loc[data2["class"]==1]
d2g=data2.loc[data2["class"]==2]
```

```python
plt.scatter(d2r['length (cm)'],d2r['width (cm)'],c="r",label="class0")
plt.scatter(d2b['length (cm)'],d2b['width (cm)'],c="b",label="class1")
plt.scatter(d2g['length (cm)'],d2g['width (cm)'],c="g",label="class2")
plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.legend()
plt.grid(True)
plt.show()
```

問3(2).png

```python
data3=pd.read_csv("https://raw.githubusercontent.com/Airi2000/Basic-Python/master/colab/quizzes/matplotlib_quiz/line_data.csv")
```

```python
plt.plot(data3["期間"],data3["件数"],marker="o")
plt.grid()
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("件数")
plt.xticks(rotation=30)
plt.show()
plt.bar(data3["期間"],data3["費用"])
plt.grid()
plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.xticks(rotation=30)
plt.show()
```
